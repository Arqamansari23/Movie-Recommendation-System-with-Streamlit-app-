# Feature Construction 

We need to add a single column named 'tag' that countain all the information about the movie genres author director and key words 

to do this we need to preprocess each column and extract information hidden in them 

### Creating Genres column 

```python
movies['genres']   # We have list of dictionaries countaining information 
```

```python
''''
def generate_genres(obj):
    genres=[]   # Creates a list that hold genres  
    for i in obj:  # loop to get all the dictionaries in a list  
        genres.append(i['name'])  # extract name of each genre  and append in the list
    return genres
  '''      
```

```python
#generate_genres(movies['genres'] )  # but the problem is the list countaining dictionaries is in string form we want to change tp list type 
```

```python
# changing to list type 

import ast  # library to convert string to list

def generate_genres(obj):
    genres=[]   # Creates a list that hold genres  
    for i in ast.literal_eval(obj):  # loop to get all the dictionaries in a list  and to convert into list type  not in str 
        genres.append(i['name'])  # extract name of each genre  and append in the list
    return genres
        
```

```python
movies['genres'].apply(generate_genres)  # we can see that we get the list of all the genres  
```

```python
movies['genres']=movies['genres'].apply(generate_genres)
```

```python
movies.head()
```

 All the unwanted values are removed from the genres column

### Creating Keyword  column 

we can apply same generate_genres at keyword also because both are in same format 

```python
movies['keywords']=movies['keywords'].apply(generate_genres)
```

### Creating Cast column 


In cast column we only need to take name of top 3 actors in the movie 

Every actor information is saved in diffrent dictionaries 

```python
movies['cast'][0]   #Every actor information is saved in diffrent dictionaries
```

```python
def actor_names(obj):
    name=[]
    counter=0   # counter to see that only 3 values can pass 
    for i in (ast.literal_eval(obj)):
            if counter!=3:  
                name.append(i['name'])
                counter=counter+1
            else:
              break
        
    return name
        
        
```

```python
 movies['cast'].apply(actor_names)
```

```python
movies['cast']=movies['cast'].apply(actor_names)
```

```python
movies.head()
```

we have Top 3 famous actors of each movies and removed all unwanted values 

### Creating Crew column 

In crew column we only want that person whose job is Director means name of director of movie 

```python
def director_names(obj):
    name=[]
    for i in (ast.literal_eval(obj)):
            if (i['job']=='Director'):  
                    name.append(i['name'])
                    break
    return name
        
        
```

```python
movies['crew'].apply(director_names)   # all the directors names are given all the outher data is removed
```

```python
movies['crew']=movies['crew'].apply(director_names)
```

```python
movies.head()
```

Only Director name is present in the crew column all the outher values are removed

# Word Vectorization 


convert words to vector array with 5000 top words

### Stemming The words of Tags to reduce size of vectors

Stemming is The Process To reduce word to their root word

Example : 

loved=love ,loving=love

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def word_stemmer(obj):
    stemmed=[]
    for i in obj.split():   # convert into list then itterate each word 
        stemmed.append(ps.stem(i))   # stem each word 
        
    return " ".join (stemmed)    # Convert list to string again 
        
        
```

```python
new['tags'].apply(word_stemmer)
```

```python
new['tags']=new['tags'].apply(word_stemmer)    # all the worlds are stemmed 
```

```python
new['tags'][0]   
```

### Words vectorization 

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english') # removing stoping_word (is am are for ) 
# max features is 5000 it picks up 5000 most frequently used words for vectorization
```

```python
cv.fit_transform(new['tags'])   # it return sparce metrix 
```

```python
# we need to convert sparce matrix into numpy array
vectors=cv.fit_transform(new['tags']).toarray()
```

```python
vectors
```

```python
vectors.shape   # we have 4806 movies and 5000 words in column as vector 
```

### Extracting Top 5000 words 

```python
cv.get_feature_names_out()    #All the 5000 top used  words
```

# Using Cosine  Similarity Funtion to find Similar Vectors 

we are using cosine similarity function to find similar vectors by finding angle between two vectors 

If the angle between two vectors is less then the two vectors are similar to each other 

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors) 
```

```python
similarity
```

```python
similarity.shape    # it is the distance of each 4806 movies with every 4806 movies 
```

# Recommendation Model 

### Feaching Movie Index 

```python
new[new['title']=='Avatar'].index[0]    # index of movie is feached 
```

### Search Highest similarity in Similarity Matrix 

Highest values indicate highest Similarity  in similarity matrix 

```python
similarity[0]     # similarity of Avatar with every outher movie 
```

```python
# sorting to get highest similarity 
sorted(similarity[0],reverse=True)   # reverse= true means Decending Order But the order of index is changed 
```

```python
# To prevent the change in order of movies indexes we are using enumerate Function 
sorted(enumerate(similarity[0]),reverse=True)  # We got the indexes but it sorted with the indexes we want sort by values
```

```python
#sorting by value 
```

```python
 sorted(list(enumerate(similarity[0])),reverse=True,key = lambda x: x[1])   # now sorted and indexes are remain same 
```

### Model Creation 

```python
def recommend_system(movie):
    index=new[new['title']==movie].index[0]   # index of movie 
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])  # all the sorted distances 
    for i in distances[1:6]: # because we need only Top 5 distances 
        print(new.iloc[i[0]].title)  # print the Title of movie 
    
```

```python
recommend_system('Avatar')  # It recomment 5 movies  which has nearest distance from avatar 
```

```python
recommend_system('Batman Begins')
```
